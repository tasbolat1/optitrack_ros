```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("hand_index.csv", header=None, names=["t", "x", "y", "z"])
```

```python
fig, ax = plt.subplots(3, figsize=(15,6))
fig.suptitle('Index finger') 
ax[0].plot(df.t, df.x, 'r-')
ax[0].set_ylabel('X')
ax[1].plot(df.t, df.y, 'g-')
ax[1].set_ylabel('Y')
ax[2].plot(df.t, df.z, 'b-')
ax[2].set_ylabel('Z')
ax[2].set_xlabel('Time, s')
plt.show()
```
